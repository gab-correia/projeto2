## 1o. TRIMESTRE - 2023

Exemplo de como ler a base de dados a partir da `url1_full`. <br>

Para saber o significado das linhas e colunas, veja o arquivo Excel disponibilizado no Blackboard nomeado: <br>
`dicionario_PNADC_microdados_trimestre1_20240527.xlsx`

```python
# Colunas selecionadas, POR EXEMPLO:
colunas = ['Ano',
           'Trimestre',
           'UF',    # Estado
           'V1022', # Situação do domicílio
           'V2007', # Sexo
           'V2009', # Idade
           'V2010', # Cor ou raça    
           'S15001'] # Tomou vacina para covid?    
```

```python
# Leitura da base de dados filtrando apenas as colunas contidas na lista `colunas`
pnad23_1_full = pd.read_csv(url1_full, usecols=colunas)
```

```python
pnad23_1_full
```

```python
pnad23_1_full.rename(columns={'V1022':'Area',
                               'V2007':'Sexo',
                               'V2009':'Idade',
                               'V2010':'CorRaca',
                               'S15001':'VacinaCovid'}, inplace=True)
pnad23_1_full
```

### IMPORTANTE ALTERAR DESCRIÇÃO DA CATEGORIA DAS VARIÁVEIS QUALITATIVAS 
Para saber o significado das categorias, veja o arquivo Excel `dicionario_PNADC_microdados_trimestre1_20240527.xlsx`

```python
# Dataframe original com categorias representadas por NÚMEROS
pnad23_1_full.Sexo.value_counts()
```

```python
# Alterando as categorias pelas suas descrições - VEJA AULA DO NOSSO CURSO
pnad23_1_full.Sexo = pnad23_1_full.Sexo.astype('category')
pnad23_1_full.Sexo = pnad23_1_full.Sexo.cat.rename_categories({1:'Homem', 2:'Mulher'})
pnad23_1_full.Sexo.value_counts(True).round(3)*100
```
